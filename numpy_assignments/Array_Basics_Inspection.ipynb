```python
import numpy as np
import requests

# Correct dataset URL
url = "https://archive.ics.uci.edu/static/public/186/wine+quality.zip"

# Download file
response = requests.get(url)
# Save zip file
with open("winequality.zip", "wb") as f:
    f.write(response.content)

print("Download completed")
```

```python
import zipfile

# Extract zip file
with zipfile.ZipFile("winequality.zip", 'r') as zip_ref:
    zip_ref.extractall()

print("Files extracted")
```

```python
data = np.genfromtxt(
    "winequality-red.csv",
    delimiter=";",
    skip_header=1
)
print(data)
```

```python
print("Shape:", data.shape)
print("Dimensions:", data.ndim)
print("Data Type:", data.dtype)
print("Total Elements:", data.size)

print("Number of wines:", data.shape[0])
print("Number of features:", data.shape[1])
```

```python
# Q1. After loading, print the shape, number of dimensions, data type, and total number of
elements in the array. How many wines and how many features are recorded?

print("Shape :", data.shape)
print("Dimensions :", data.ndim)
print("Data Type :", data.dtype)
print("Total Elements :", data.size)
print("Number of wines :", data.shape[0])
print("Number of features :", data.shape[1])
```

```python
# Q2. Print the first 5 rows and the last 3 rows. Then print only
# the alcohol and quality columns for the first 10 rows.
print("First 5 Rows:")
print(data[:5])
print("\nLast 3 Rows:")
print(data[-3:])
print("\nAlcohol and Quality Columns (First 10 Rows):")
print(data[:10, [10, 11]])
```

```python
# Q3. What is the minimum and maximum value in the entire dataset? Which column has the
# largest range (max - min)? Compute this without a loop.
# Find minimum value in entire dataset
print("Minimum Value:", np.min(data))

# Find maximum value in entire dataset
print("Maximum Value:", np.max(data))

# Find range of each column
column_range = np.max(data, axis=0) - np.min(data, axis=0)
print("\nRange of Each Column:")
print(column_range)

# Find column index with largest range
largest_range_column = np.argmax(column_range)

print("\nColumn with Largest Range:", largest_range_column)
print("Largest Range Value:", column_range[largest_range_column])
```

```python
# Q4. Check if there are any NaN values in the dataset. Print the count of NaN values per
# column. If there are none, explain in a markdown cell how np.genfromtxt handles missing
# entries by default and what you would do to handle them.
print("Are there any NaN values?")
print(np.isnan(data).any())
print("\nNaN values in each column:")
print(np.isnan(data).sum(axis=0))
```

```python
# Q5. Extract the quality column (column index 11) into a separate 1D array called quality.
# Find all unique quality ratings present and how many wines belong to each rating - using only NumPy.

# Extract the quality column (index 11)
quality = data[:, 11]

# Find unique quality ratings and their counts
unique_values, counts = np.unique(quality, return_counts=True)

print("Quality Column:")
print(quality)
print("\nUnique Quality Ratings:")
print(unique_values)
print("\nQuality Rating Counts:")
for i in range(len(unique_values)):
    print(f"Quality {unique_values[i]} = {counts[i]} wines")
```
